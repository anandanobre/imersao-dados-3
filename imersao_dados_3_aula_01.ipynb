## Visualizção e análise dos dados

```python
# Importa biblioteca e le o arquivo
import pandas as pd 

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

df = pd.read_csv(url_dados, compression = 'zip')
display(df)
```

```python
# Mostra 5 primeiras linhas da tabela
df.head()
```

```python
# Mostra quantidade de linhas e colunas da tabela
df.shape
```

```python
# Retorna as categorias da coluna
print("Tratamento: ", df["tratamento"].unique())
print("Tempo: ", df["tempo"].unique())
print("Dose: ", df["dose"].unique())
print("Droga: ", df["droga"].unique())
```

```python
df["g-0"].unique()
```

#### Com os dados observados até agora sabemos que:
   - O id é o identificador para cada cell experimentada.
   - O tratamento possui dois tipos, onde o 'com_controle' significa garantir que o experimento está sob as mesmas condições com excessão da variável estudada, dessa forma podendo afirmar que os resultado saó de fato devido ao experimento.
   - O tempo é a duração de cada experimento, o tempo de exposição.
   - A dose são as variações de quantidade usada da droga. 
   - A droga é a própria substância usada.
   - Os g- e c- são a expressão(quantidade e RNA produzido) de cada gene para cada droga e quantidade submetida
