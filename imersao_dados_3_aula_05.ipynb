```python
# Importando base de dados completa
import pandas as pd
url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

df_dados = pd.read_csv(url_dados, compression = 'zip')
df_dados.head()
```

```python
# Importa basede dados dos resultados
url_dados_resultados = 'https://github.com/mjuliacsouza/imersaodados3/blob/main/dados/dados_resultados.csv?raw=true'

df_result = pd.read_csv(url_dados_resultados)
df_result.head()
```

```python
#usando o drop para retirar a coluna de id
contagem_drop = df_result.drop('id', axis=1).sum().sort_values(ascending=False)
contagem_drop
```

```python
# Contagem por linhas
cont_row = df_result.drop('id', axis=1).sum(axis=1)
cont_row
```

```python
# Criando a coluna da soma de moas por linhas
df_result['n_moa'] = cont_row
df_result['ativo_moa'] = df_result['n_moa'] != 0
df_result.head()
```

```python
# Fazendo merge das duas tabelas
df_merged = pd.merge(df_dados, df_result[['id','n_moa','ativo_moa']], on='id')
df_merged
```

## Criando moledos preditivos

#### Modelo de Logitic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Separando dados de treino e de teste
x = df_merged.select_dtypes('float64')
y = df_merged['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

modelo_RL = LogisticRegression(max_iter=1000)
# Treinando modelo
modelo_RL.fit(x_treino, y_treino)
# Calcula a acurácia do modelo
modelo_RL.score(x_teste, y_teste)
```

#### Modelo Dummy
Equivalete a chutar que todos os casos houve ativação 

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

#### Modelo Decison Tree

```python
from sklearn.tree import DecisionTreeClassifier

x = df_merged.select_dtypes('float64')
y = df_merged['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

modelo_arvore = DecisionTreeClassifier(max_depth=3)
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)
```

```python
# Melhorando o modelo de Decision tree

x = df_merged.select_dtypes('float64')
y = df_merged['ativo_moa']
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

teste = []
treino = []

for i in range(1, 15):
    modelo_arvore = DecisionTreeClassifier(max_depth=i)
    modelo_arvore.fit(x_treino, y_treino)
    teste.append(modelo_arvore.score(x_teste, y_teste))
    treino.append(modelo_arvore.score(x_treino, y_treino))
```

```python
# Visualizando gráfico para comparação
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
sns.lineplot(x=range(1,15), y = teste, label='teste')
sns.lineplot(x=range(1,15), y = treino, label='treino')
```
