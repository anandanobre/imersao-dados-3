```python
# Importa biblioteca e le o arquivo
import pandas as pd 

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

df = pd.read_csv(url_dados, compression = 'zip')
```

```python
# Altera o nome da coluna
mapa = {'droga': 'composto'}
df.rename(columns = mapa, inplace = True)

df.head()
```

## Correlacionando os dados

```python
# Cria tabela de correlaÃ§Ã£o
pd.crosstab([df['dose'], df['tempo']], df['tratamento'])
```

```python
pd.crosstab([df['dose'], df['tempo']], df['tratamento'], normalize=True)
```

```python
pd.crosstab([df['dose'], df['tempo']], df['tratamento'], normalize='index')
```

```python
pd.crosstab([df['dose'], df['tempo']], df['tratamento'], values = df['g-0'], aggfunc='mean')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()

sns.scatterplot(x='g-0', y='g-3', data = df)
plt.show()
```

```python
sns.scatterplot(x='g-0', y='g-8', data = df)
plt.show()
```

```python
sns.lmplot(data = df, x='g-0', y='g-8', line_kws={'color': 'red'})
plt.show()
```

```python
sns.lmplot(data = df, x='g-0', y='g-8', line_kws={'color': 'red'}, col='tratamento', row='tempo')
plt.show()
```

```python
df.loc[:, 'g-0': 'g-771'].corr()
```

```python
import numpy as np

# Compute the correlation matrix
corr = df.loc[:, 'g-0': 'g-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Compute the correlation matrix
corr_cell = df.loc[:, 'c-0': 'c-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_cell, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_cell, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

### Desafio 05

```python
#Correlacionando G e C

# Compute the correlation matrix
corr_cell = pd.concat([df.loc[:,'g-0':'g-9'], df.loc[:,'c-0':'c-9']], axis=1)
cor = corr_cell.corr().loc['c-0':'c-9','g-0':'g-9']

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(cor,  cmap=cmap, center=0,   
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```
