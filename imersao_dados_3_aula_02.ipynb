```python
# Importa biblioteca e le o arquivo
import pandas as pd 

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

df = pd.read_csv(url_dados, compression = 'zip')
```

```python
# Altera o nome da coluna
mapa = {'droga': 'composto'}
df.rename(columns = mapa, inplace = True)

df.head()
```

```python
# Selecionando os nomes dos 5 compostos que aparecem com mais frequência
comp_list = df['composto'].value_counts().index[:5]
comp_list
```

```python
# Filtrando dados com query
# O @ serve para dizer que o valor está vindo de fora da string
df.query('composto in @comp_list')
```

## Plotando dados

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set() # Salva as configurações de imagens padrão do seaborn
plt.figure(figsize = (8, 8))
fig = sns.countplot(x = 'composto', data = df.query('composto in @comp_list'))
fig.set_title('Top 5 compostos')
plt.show
```

```python
# Quantidades de elementos distintos no g-0
len(df['g-0'].unique())
```

```python
# Pega os menor valor e o maior valor contido em g-0
print(df['g-0'].min())
print(df['g-0'].max())
```

```python
# Plotando histograma

fig, ax = plt.subplots(1,2, figsize = ( 20 , 4 ))

df['g-0'].hist(bins = 100, ax = ax[0])
df['g-19'].hist(bins = 100, ax = ax[1])
```

```python
# Descreve os dados com algumas estátiscas
df.describe()
```

```python
# Faz slicing de duas colunas
df[['g-0', 'g-1']]
```

```python
# Faz slice de todos os g-
df.loc[:, 'g-0': 'g-771'].describe()
```

```python
# Faz a tabela transposta
df.loc[:, 'g-0': 'g-771'].describe().T
```

```python
# Faz histograma da média, dos mínimos e dos máximos dos g-

fig, ax = plt.subplots(1,3, figsize = ( 20 , 4 ))

# Histograma da média
df.loc[:, 'g-0': 'g-771'].describe().T['mean'].hist(bins = 30, ax = ax[0]).set_title('Médias', fontsize = 13)

# Histograma do minímo
df.loc[:, 'g-0': 'g-771'].describe().T['min'].hist(bins = 30, ax = ax[1]).set_title('Mínimos', fontsize = 13)

# Histograma do máximo
df.loc[:, 'g-0': 'g-771'].describe().T['max'].hist(bins = 30, ax = ax[2]).set_title('Máximos', fontsize = 13)

plt.show()
```

```python
# Faz histograma da média, dos mínimos e dos máximos dos c-

fig, ax = plt.subplots(1,3, figsize = ( 20 , 4 ))

# Histograma da média
df.loc[:, 'c-0': 'c-99'].describe().T['mean'].hist(bins = 30, ax = ax[0]).set_title('Médias', fontsize = 13)

# Histograma do minímo
df.loc[:, 'c-0': 'c-99'].describe().T['min'].hist(bins = 30, ax = ax[1]).set_title('Mínimos', fontsize = 13)

# Histograma do máximo
df.loc[:, 'c-0': 'c-99'].describe().T['max'].hist(bins = 30, ax = ax[2]).set_title('Máximos', fontsize = 13)

plt.show()
```

```python
# Plotando gráfico boxplot

plt.figure(figsize = (8, 8))
sns.boxplot( y = 'g-0', x = 'tratamento', data = df)
plt.show()
```
